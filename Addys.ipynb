Now to run with all hospitals

```python
addys = []
def get_addys():
    for i in range(0, len(location)):
        url = base_url+location[i]+end_url
        response= get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        info = soup.find('div',{'class':'hp-details-address'}).text
        addys.append(info)
        
    return addys
    
get_addys()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=['location','address'])
```

```python
df.location = location
df.address = addys
```

```python
df.address = df.address.str.strip()
```

```python
df.location = df.location.str.replace('-', ' ')
df.location = df.location.str.replace('/', '')
```

```python
df
```

```python
df['address'] = df['address'].replace(to_replace=r'(Street|Avenue|Ave|Road|Parkway|Drive|Blvd|Way|Highway)', value=r'\1 ', regex=True)
```

```python
df['address'] = df['address'].replace(to_replace=r'(\d{5})', value=r'\1 ', regex=True)
```
