```python
def get_hosp_data(hosp_list):
    baseurl = 'https://oregonhospitalguide.org/hospitals/'
    hospital_names = []
    prices = []
    for i in hosp_list:
        url = requests.get(baseurl + i).text
        hosp_soup = BeautifulSoup(url, "lxml")
        proceds = hosp_soup.find_all('li',{'class':'cp-procedure'})
        hprocs = [proceds[i].contents[1].text[:-2] for i in range(len(proceds))]
        if len(hprocs)<81:
            hospital_names.append(hosp_soup.title.text[:-17])
            proc_price = np.zeros(80)
            for i in range(len(hprocs)):
                if proceds[i].contents[1].text[:-2] in proc_lookup.values():
                    proc_price[rev_lu[proceds[i].contents[1].text[:-2]]] = proceds[i].contents[5].text[1:].replace(',','')
            prices.append(proc_price)
            
    return hospital_names, prices
```
